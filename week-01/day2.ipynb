# ðŸ”¹ 1. Variables and Data Types

Variables are used to store values. Python is dynamically typed, meaning you donâ€™t need to declare the type explicitly.

```python
# Variables and Data Types
x = 10          # int
y = 3.14        # float
name = "Fayaz"  # string
is_active = True # boolean

print("x:", x, type(x))
print("y:", y, type(y))
print("name:", name, type(name))
print("is_active:", is_active, type(is_active))
```

# ðŸ”¹ 2. Operators with Examples

Operators allow us to perform operations on variables.

```python
# Arithmetic Operators
a, b = 15, 4
print(f" a = {a} and b = {b}")
print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)  # division
print("a % b =", a % b)  # modulus
print("a ** b =", a ** b) # power
print("a // b =", a // b) # floor division

# Comparison Operators
print("a == b:", a == b)
print("a > b:", a > b)

# Logical Operators
print("a > 10 and b < 10:", a > 10 and b < 10)

# Membership Operator
print("'F' in 'Fayaz':", 'F' in "Fayaz")

# Identity Operator
x = [1, 2, 3]
y = x
print("x is y:", x is y)
```

# ðŸ”¹ 3. Strings in Depth

Strings are sequences of characters.

```python
# String indexing and slicing
s = "Python Programming"
print("First character:", s[0])
print("Last character:", s[-1])
print("Slice [0:6]:", s[0:6])

# String methods
print("Uppercase:", s.upper())
print("Lowercase:", s.lower())
print("Split:", s.split())
print("Replace:", s.replace("Python", "AI"))

# String formatting
name = "Fayaz"
age = 25
print(f"My name is {name} and I am {age} years old.")  # f-string

```

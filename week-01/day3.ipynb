# ğŸ”¹ 1. Lists

Definition: A list is an ordered, mutable (changeable) collection.

Properties:

Allows duplicates

Indexed (can access by position)

Supports many built-in methods (append, remove, sort, etc.)

Use case: When you need an ordered collection that can be modified.

```python
# LIST Example
fruits = ["apple", "banana", "mango"]
print("Original List:", fruits)

# Add item
fruits.append("orange")
print("After append:", fruits)

# Access by index
print("First fruit:", fruits[0])

# Modify
fruits[1] = "grapes"
print("After modification:", fruits)

# Lists allow duplicates
fruits.append("apple")
print("With duplicate:", fruits)
```

# ğŸ”¹ 2. Tuples

Definition: A tuple is an ordered, immutable collection.

Properties:

Indexed (like lists)

Cannot be modified after creation

Faster than lists (because of immutability)

Use case: When you want fixed data (e.g., coordinates, constant values).

```python
# TUPLE Example
colors = ("red", "green", "blue")
print(colors)
print("Tuple:", colors)

# Access by index
print("First color:", colors[0])

# Tuples cannot be modified
# colors[1] = "yellow"  # âŒ This will cause an error

# But you can have duplicates
colors = ("red", "green", "red")
print("Tuple with duplicate:", colors)
```

# ğŸ”¹ 3. Sets

Definition: A set is an unordered collection of unique elements.

Properties:

No duplicates allowed

No indexing (cannot access elements by position)

Supports set operations like union, intersection, difference

Use case: When uniqueness is important (e.g., removing duplicates).

```python
# SET Example
nums = {1, 2, 3, 3, 4}
print("Set removes duplicates automatically:", nums)

# Add element
nums.add(5)
print("After adding:", nums)

# Remove element
nums.remove(2)
print("After removing:", nums)

# Set operations
a = {1, 2, 3}
b = {3, 4, 5}
print("Union:", a | b)         # or a.union(b)
print("Intersection:", a & b)  # or a.intersection(b)
print("Difference:", a - b)
```

# ğŸ”¹ 4. Dictionaries

Definition: A dictionary stores key-value pairs.

Properties:

Unordered (Python 3.7+ keeps insertion order)

Keys must be unique

Values can be duplicated

Keys must be immutable (string, number, tuple)

Use case: When you need to map data (like a studentâ€™s name â†’ grade).

```python
# DICTIONARY Example
student = {"name": "Fayaz", "age": 25, "grade": "A"}
print("Original Dictionary:", student)

# Access by key
print("Name:", student["name"])

# Modify value
student["age"] = 26
print("After modification:", student)

# Add new key-value pair
student["city"] = "Peshawar"
print("After adding new key:", student)

# Dictionary methods
print("Keys:", student.keys())
print("Values:", student.values())
```

# ğŸ”¹ 1. If / Else Statements

Definition: Conditional statements allow you to execute certain blocks of code only if a condition is true.

Keywords: if, elif (else if), else

Use case: Decision making (checking conditions).

```python
# IF / ELSE Example
num = 15

if num > 0:
    print("Positive number")
elif num == 0:
    print("Zero")
else:
    print("Negative number")

# Nested If Example
age = 20
if age >= 18:
    if age < 21:
        print("You are an adult but not 21 yet.")
    else:
        print("You are 21 or older.")
else:
    print("You are under 18.")
```

# Definition:
 A for loop is used for iterating over a sequence (list, string, tuple, dictionary, etc.).

Properties:

Iterates element by element

Uses the keyword for

Use case: When you know the number of iterations.

```python
# FOR LOOP Example
fruits = ["apple", "banana", "mango"]

for fruit in fruits:
    print("Fruit:", fruit)

# Iterating through a string
for char in "Python":
    print("Character:", char)

# Using range()
for i in range(1, 6):
    print("Number:", i)
```

# ğŸ”¹ 4. Loop Control Statements

break â†’ Exits the loop immediately

continue â†’ Skips current iteration, moves to the next

pass â†’ Does nothing (placeholder)

```python
# LOOP CONTROL Example
for i in range(1, 6):
    if i == 3:
        continue   # skip number 3
    if i == 5:
        break      # stop loop when 5
    print("Number:", i)

# Using pass
for i in range(3):
    pass  # placeholder for future code
```

# 3. While Loop

Definition: A while loop runs as long as a condition is true.

Properties:

Needs a condition to continue looping

Can run infinitely if not handled correctly

Use case: When you donâ€™t know the number of iterations in advance.

```python
# Simple WHILE LOOP Example
count = 1

while count <= 5:
    print("This is loop number:", count)
    count += 1  # increase count each time

print("Loop finished!")
```

```python
# WHILE LOOP Example
count = 0

while count <= 5:
    print("Count:", count)
    count += 1  # update condition to avoid infinite loop

# Example: user input until correct
secret = "python"
guess = ""
while guess != secret:
    guess = input("Guess the secret word: ")
print("Correct! The word is", secret)
```

# Functions â€” simple explanation

A function is a reusable block of code that performs one task. You define a function once, then call it whenever you need that task done. Functions help you organize code, avoid repetition, and make programs easier to read and test.

```python
# Cell 1: Basic function example
def greet(name):
    """Return a greeting for the given name (this is a docstring)."""
    return f"Hello, {name}!"

# call the function
greet("Fayaz")   # Output: Hello, Fayaz!
```

```python
# Cell 2: Parameters and arguments
def introduce(name, age, city="Peshawar"):
    """Return a short introduction. 'city' has a default value."""
    return f"{name} is {age} years old and lives in {city}."

# Positional arguments
print(introduce("Ali", 30))

# Keyword arguments (order doesn't matter)
print(introduce(age=22, name="Sara"))

# Override default value
print(introduce("Fayaz", 25, city="Islamabad"))
```

```python
from os import lstat
# Cell 3: Mutable default pitfall (don't do this in practice)
def add_item_bad(item, lst=[]):
    """Bad: default list is shared across calls."""
    lst.append(item)
    return lst

print(add_item_bad(1))  # [1]
print(add_item_bad(2))  # [1, 2]  <-- surprising: same list reused

# Correct pattern
def add_item(item, lst=None):
    """Good: create a new list when None is passed."""
    if lst is None:
        lst = []
    lst.append(item)
    return lst

print(add_item(1))      # [1]
print(add_item(2))      # [2] (separate list)
```

ğŸŸ¢ 1. What are Modules?

A module is simply a Python file (.py) that contains code (functions, classes, or variables) which you can import and reuse in other programs.

Think of modules as toolboxes:

Built-in modules (already in Python, e.g., math, os, random).

User-defined modules (files you create yourself).

```python
import math

print(math.sqrt(16))    # 4.0
print(math.pi)          # 3.141592653589793
```

```python
from random import randint

print(randint(1, 10))   # Random number between 1 and 10
```
